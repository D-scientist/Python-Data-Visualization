# Hue Colour Change 

```python
#STEP-1 import laiberaries 
import  pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# STEP-2 Load Data Set 
tip = sns.load_dataset("tips")
tip.head(5)
```

```python
sns.boxplot(x='tip',y='day',hue='smoker',data=tip)
```

```python
sns.boxplot(x='tip',y='day',hue='smoker',data=tip, palette={'Yes': 'tab:blue', 'No': 'tab:green',} ,saturation=2)
```

Lets Take Another example from  Iris Data Set

```python
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid", palette="muted")

# Load the example iris dataset
iris = sns.load_dataset("iris")
iris.columns
# "Melt" the dataset to "long-form" or "tidy" representation
iris = pd.melt(iris, "species", var_name="measurement")

# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="measurement", y="value", hue="species", data=iris)
```

```python
sns.set(style="whitegrid", palette="muted")
iris = sns.load_dataset("iris")
iris = pd.melt(iris, "species", var_name="measurement")
sns.swarmplot(x="measurement", y="value", hue="species", data=iris, palette=dict(setosa = 'g', versicolor = 'b', virginica = 'r'))
```
